```
#Esta función recibe 3 parámetros, año: string, csv_separator:string se refiere al separador del csv, has_header:true or false se refiere a si el CSV tiene o no header.
def read_file_by_year(year, csv_separator, has_header):
  file_path = "/mnt/storage/datos_input/Titulados_Educacion_Superior_" + year + ".csv"
  return spark.read.format("csv").option("sep", csv_separator).option("inferSchema", "true").option("header", has_header).load(file_path)
```

```
#readFilesHasDataFrame: recibe 2 parámetros, _years:array de string y _dataFrames:array esta variable es donde se van a guardar los dataframes
def readFilesHasDataFrame(_years, _dataFrames):
  for year in _years:
    _dataFrames.append(read_file_by_year(year, ";", "true"))
    
yearArray = ["2018", "2019"]
dataFrames = []
readFilesHasDataFrame(yearArray, dataFrames)
```

```
def showDataFrame(arrayDataFrame):
  for dataFrame in arrayDataFrame:
    dataFrame.show()
showDataFrame(dataFrames)
```

```
#createParquetByDaframes: recibe 3 parámetros _partitionBy:string se refiere al campo de la partición, _path: string se refiere al path donde se va a generar el parquet, _arrayDataFrame: array que contiene los Dataframes a particionar.
def createParquetByDaframes(_partitionBy, _path, _arrayDataFrame): 
  for dataFrame in _arrayDataFrame:
    if _arrayDataFrame.index(dataFrame) != 0:
      mode = "append"
    else:
      mode = "overwrite"
    dataFrame.write.format("parquet").mode(mode).partitionBy(_partitionBy).save(_path)
```

```
parquetFilePath = "/mnt/storage/datos_input/lab3.parquet"
createParquetByDaframes("CAT_PERIODO", parquetFilePath, dataFrames)
```

```
# Gráfico que muestre la cantidad de graduados por cada institución de educación para el año 2019:
grafico_uno = spark.sql('''SELECT COUNT(NOMB_INST) AS CANTIDAD_DE_GRADUADOS, NOMB_INST  AS INSTITUCION_DE_EDUCACION  FROM titulados_educacion_table_view  WHERE SUBSTRING(FECHA_OBTENCION_TITULO, 1, 4) = "2019" GROUP BY NOMB_INST ORDER BY NOMB_INST DESC''')
display(grafico_uno)
```

```
# Gráfico que muestre la cantidad de graduados por cada carrera para la institución de educación “IP DUOC UC”, comparando los años 2018 y 2019:
grafico_dos = spark.sql('''SELECT COUNT(NOMB_CARRERA) AS CANTIDAD_DE_GRADUADOS, NOMB_CARRERA AS CARRERA, SUBSTRING(FECHA_OBTENCION_TITULO, 1, 4) AS AGNO FROM titulados_educacion_table_view  WHERE NOMB_INST = "IP DUOC UC" AND (SUBSTRING(FECHA_OBTENCION_TITULO, 1, 4) = "2019" OR SUBSTRING(FECHA_OBTENCION_TITULO, 1, 4) = "2018") GROUP BY NOMB_CARRERA, AGNO''')
display(grafico_dos)
```

```
#Gráfico que muestre la evolución de egresados desde el año 2018 a 2019, comparando las Jornadas diurnas y vespertinas:
grafico_tres = spark.sql('''SELECT COUNT(JORNADA) AS EGRESADOS, SUBSTRING(FECHA_OBTENCION_TITULO, 1, 4) AS AGNO, JORNADA FROM titulados_educacion_table_view  WHERE (SUBSTRING(FECHA_OBTENCION_TITULO, 1, 4) = "2019" OR SUBSTRING(FECHA_OBTENCION_TITULO, 1, 4) = "2018") AND JORNADA IN ("VESPERTINO", "DIURNO") GROUP BY JORNADA, AGNO ORDER BY AGNO ASC''')
display(grafico_tres)
```
